Lo que representan las varibles son lo siguiente: la variable City
indica la ubicación en cuestión donde se analiza el ritmo de vida de 
las personas. Walk indica la rapidez media de un peatón al caminar.
Bank representa el tiempo promedio de un empleado para dar cambio.
Talk representa la rapidez promedio de conversación de un empleado 
postal. Heart indica la tasa de mortalidad por edad causada por
enfermedad cardiaca.

```python
print("EL rango de Walk es de",datapf['Walk'].min(),
      " a",datapf['Walk'].max())

print("EL rango de Bank es de",datapf['Bank'].min(),
      " a",datapf['Bank'].max())

print("EL rango de Talk es de",datapf['Talk'].min(),
      " a",datapf['Talk'].max())
print("EL rango de Heart es de",datapf['Heart'].min(),
      " a",datapf['Heart'].max())
```

```python
print("La media de Walk es",datapf['Walk'].mean(),".",
      "La mediana de Walk es",datapf['Walk'].median(),".",
      "La desviación estándar de Walk es",datapf['Walk'].std(),".")

print("La media de Bank es",datapf['Bank'].mean(),".",
      "La mediana de Bank es",datapf['Bank'].median(),".",
      "La desviación estándar de Bank es",datapf['Bank'].std(),".")

print("La media de Talk es",datapf['Talk'].mean(),".",
      "La mediana de Talk es",datapf['Talk'].median(),".",
      "La desviación estándar de Talk es",datapf['Talk'].std(),".")

print("La media de Heart es",datapf['Heart'].mean(),".",
      "La mediana de Heart es",datapf['Heart'].median(),".",
      "La desviación estándar de Heart es",datapf['Heart'].std(),".")
```
